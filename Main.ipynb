```python
import pandas as pd

file_path = 'AffiliatedStore.csv'
data = pd.read_csv(file_path, encoding='cp949')

print(data.head())
```

```python
category_counts = data['Category'].value_counts()
print(category_counts)
```

```python
type_counts = data['Type'].value_counts()
print(type_counts)
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
!apt-get update -qq
!apt-get install -y fonts-nanum

# 설치된 폰트를 matplotlib에 추가
fm.fontManager.addfont('/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf')

# 기본 폰트를 나눔 폰트로 설정
plt.rc('font', family='NanumBarunGothic')

offline_data = data[data['Type'] != '온라인']

category_counts_offline = offline_data['Category'].value_counts()

plt.figure(figsize=(10, 6))
category_counts_offline.plot(kind='bar')
plt.title('가맹점 분류', fontproperties=fm.FontProperties(fname='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'))
plt.xlabel(' ', fontproperties=fm.FontProperties(fname='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'))
plt.ylabel('가맹점 수', fontproperties=fm.FontProperties(fname='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'))
plt.xticks(fontproperties=fm.FontProperties(fname='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'))
plt.yticks(fontproperties=fm.FontProperties(fname='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'))
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import squarify

# 한글 폰트 설정
!apt-get update -qq
!apt-get install -y fonts-nanum

# 설치된 폰트를 matplotlib에 추가
fm.fontManager.addfont('/usr/share/fonts/truetype/nanum/NanumGothic.ttf')

# 기본 폰트를 나눔 폰트로 설정
plt.rc('font', family='NanumGothic')

offline_data = data[data['Type'] != '온라인']

category_counts_offline = offline_data['Category'].value_counts()

# 컬러 세팅
colors = plt.cm.tab20(range(len(category_counts_offline)))

fig, ax = plt.subplots(figsize=(12, 8))
squarify.plot(sizes=category_counts_offline.values, label=category_counts_offline.index, alpha=.8, color=colors, ax=ax,
              text_kwargs={'fontsize': 12, 'fontproperties': fm.FontProperties(fname='/usr/share/fonts/truetype/nanum/NanumGothic.ttf')})

plt.axis('off')
plt.title('가맹점 분류 2', fontproperties=fm.FontProperties(fname='/usr/share/fonts/truetype/nanum/NanumGothic.ttf'))
plt.show()
```
